```python
# find the most common pair

def get_pairs(tokens):
    dic = {}
    for pair in zip(tokens, tokens[1:]):
        dic[pair] = dic.get(pair, 0) + 1
    return dic

dic = get_pairs(tokens)
sorted_dic = sorted(((v, k) for k, v in dic.items()), reverse=True)
sorted_dic
```

```python
def combine(list, pair, number):
    new_list = []

    i = 0
    while i < len(list):
        if list[i] == pair[0] and list[i + 1] == pair[1]:
            new_list.append(number)
            i += 2
        else:
            new_list.append(list[i])
            i += 1 

    return new_list

len(combine(tokens, (116, 104), 1000))
```

```python
combos = {}
```

```python
def encode(tokens, vocab_size):
    num_combinations = vocab_size - 256

    new_list = list(tokens) # copy the list

    for i in range(num_combinations):
        # calculate the current number
        num = i + 256
        
        # get the max number of pairs
        max_pair = max(get_pairs(new_list), key=get_pairs(new_list).get)
        print(f"step {i}: merging {max_pair} into {num}")
        combos[max_pair] = num

        # change the pair to new number
        new_list = combine(new_list, max_pair, num)

    return new_list
```

```python
def decode(list):
    vocab = {idx: bytes([idx]) for idx in range(256)}
    for (p0, p1), idx in combos.items():
        vocab[idx] = vocab[p0] + vocab[p1]

    for pair, idx in vocab.items():
        print(pair, idx)

    toks = b"".join(vocab[idx] for idx in list) # text in byte literal
    return toks.decode("utf-8", errors="replace") # without replace invalid start bit
```

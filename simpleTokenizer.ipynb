```python
# find the most common pair

def get_pairs(tokens):
    dic = {}
    for pair in zip(tokens, tokens[1:]):
        dic[pair] = dic.get(pair, 0) + 1
    return dic

dic = get_pairs(tokens)
sorted_dic = sorted(((v, k) for k, v in dic.items()), reverse=True)
sorted_dic
```

```python
def combine(list, pair, number):
    new_list = []

    i = 0
    while i < len(list):
        if list[i] == pair[0] and list[i + 1] == pair[1]:
            new_list.append(number)
            i += 2
        else:
            new_list.append(list[i])
            i += 1 

    return new_list

len(combine(tokens, (116, 104), 1000))
```
